#**Import all libraries**

```python
df.value_counts().sum()
```

```python
df['gender'].value_counts()
```

```python
df = df[df['gender'] != 'Other']
```

```python
df['gender'] = np.where(df['gender'] == 'Male', 1, 2)
```

```python
df = df.drop(['smoking_history'], axis=1)
```

```python
df['gender'].value_counts()
```

#**Split dataset into train test**

```python
df.head()
```

```python
df.rename(columns = {'gender':'Gender', 'age': 'Age', 'hypertension':'Hypertension', 'heart_disease':'HeartDisease', 'bmi':'BMI','HbA1c_level':'HBA', 'blood_glucose_level': 'Glucose', 'diabetes':'Diabetes'}, inplace=True)
```

```python
df.head()
```

```python
x = df.iloc[:,0:-1]
y = df.iloc[:,-1]
```

```python
x.head()
```

```python
y.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=40)
```

```python
y_train.head()
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.fit_transform(x_test)
```

##**K-nearest neighbor Algorithm**

```python
tree_model = KNeighborsClassifier(n_neighbors=11, p=2, metric='euclidean')
ml_tree_model = tree_model.fit(x_train_scaled, y_train)
```

```python
y_predicted = ml_tree_model.predict(x_test_scaled)
```

```python
tree_accuracy = accuracy_score(y_predicted, y_test)
```

```python
tree_accuracy
```

###K-nearest neighbors algorithm give **0.9636** accuracy

##**Logistic regression algorithm**

```python
lr_model = LogisticRegression()
ml_lr_model = lr_model.fit(x_train_scaled, y_train)
```

```python
y_predicted = ml_lr_model.predict(x_test_scaled)
lr_accuracy = accuracy_score(y_predicted, y_test)
lr_accuracy
```

###Logistic regression algorithm give **0.9593** accuracy

##**Support Vector Machine algorithm**

```python
svc_model = SVC()
ml_svc_model = svc_model.fit(x_train_scaled, y_train)
```

```python
y_predicted = ml_svc_model.predict(x_test_scaled)
svc_accuracy = accuracy_score(y_predicted, y_test)
svc_accuracy
```

###Support vector machine give **0.9635** accuracy

##**Decision Trees algorithm**

```python
tree_model = DecisionTreeClassifier()
ml_tree_model = tree_model.fit(x_train_scaled, y_train)
```

```python
y_predicted = ml_tree_model.predict(x_test_scaled)
tree_accuracy = accuracy_score(y_predicted, y_test)
tree_accuracy
```

###Decision tree classifier algorithm give **0.9498** accuracy
